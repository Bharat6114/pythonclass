```python
def factorial(num):
    if(num<0):
        print("please enter positive number")
    elif(num==1):
        return 1
    elif(num==2):
        return 2
    else:
        for i in range(2,num):
            fact=num*factorial(num-1)
            return fact
        
```

```python
factorial(5)
```

```python
def unique(list1):
    list2=[]
    for i in list1:
        if i not in list2:
            list2.append(i)
        else:
            pass
    return list2
```

```python
unique([1,2,3,4,55,6,67,78,2,3,4,4])
```

```python
def prime(num):
    if(num==1):
        print("the number is prime")
    elif(num==2):
        print("the number is not prime")
    else:
        for i in range(2,num):
            if(num%i==0):
                return False
        else:
            return True
```

```python
prime(11)
```

```python
def perfect(num):
    sum=0
    for i in range(1,num):
        if(num%i==0):
            sum=sum+i
    return sum==num
```
