```python
for i in range(2,20):
    for j in range(2,11):
        print(f"{i}x{j}={i*j}")
    print(",..................")
```

```python
for i in range(2,20):
    for j in range(2,11):
        print("{}x{}={}".format(i,j,i*j))
    print(",..................")
```

```python
for i in range(0,10):
    for j in range(0,i):
        print("*",end=" ")
    print("\n")
```

```python
for i in range(0,10):
    for j in range(0,i+1):
        print(i,end=" ")
    print("\n")
```

```python
for i in range(10,-1,-1):
    for j in range(0,i):
        print("*",end=" ")
    print("\n")
```

```python
for i in range(10,-1,-1):
    for j in range(0,i+1):
        print(i,end=" ")
    print("\n")
```
